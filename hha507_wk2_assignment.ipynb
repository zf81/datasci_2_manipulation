```python
# Load packages

import pandas as pd
import numpy as np
import re
```

## 1. Data Cleaning with Pandas:

```python
# Load the dataset

df = pd.read_csv('/content/healthcare_data_cleaning.csv', na_values=['missing'])
df
```

```python
# Identify and handle missing values in the dataset

Missing_values = df.isnull()
Missing_values = Missing_values.sum().sum()
print('Missing values: ', Missing_values)
```

```python
# Using df.fillna (.fillna method) to fill the missing values as NaN

missing_values = df.fillna(np.nan)
missing_values
```

Remove any duplicate rows and columns, if they exist.

```python
 # we can use the .drop_duplicates(inplace = True) function to remove duplicate rows

df.drop_duplicates(inplace=True)
df
```

```python
# To remove duplicate columns, I used df.loc[:, ~df.columns.duplicated()]
# I found the function here- https://www.geeksforgeeks.org/how-to-find-drop-duplicate-columns-in-a-pandas-dataframe/

df = df.loc[:, ~df.columns.duplicated()]
df
```

Clean column names

```python
# I started to clean column names by removing white space using the .str.strip() method

df.columns= df.columns.str.strip()
df
```

```python
# Function to remove white space and special characters from a value
def clean_column_names(df):
      # Define a helper function to clean column names
  def clean_name(value):
    cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', value)
    return cleaned_name.lower()
  df.columns = [clean_name(col) for col in df.columns]
  return df

# Apply the clean_value function to all columns
df = clean_column_names(df)
```

```python
# Display cleaned data
df
```

## 2. Data Transformation:


```python
# Looking at the blood pressures of patients of various ages and determining if their blood pressure is normal or high

# Define clinically defined thresholds for categorization
threshold_hypertension = 140

# Create a new column "Blood_Pressure_Stage" based on thresholds
df['Blood_Pressure_Stage'] = df['averagebp'].apply(lambda x: 'High BP' if x >= threshold_hypertension else 'Normal BP')

print(df[['patientage', 'averagebp', 'Blood_Pressure_Stage']].head(10))
```

Aggregate data using groupby and compute summary statistics

```python
# Summary statistics of the visitdurationmins column

df.groupby(['visitdurationmins']).describe()
```

```python
# Examine the frequency of each disease diagnosed (flu, allergy, COVID-19) using .value_counts()

df['diseasediagnosed'].value_counts()
```

```python
# Examine the frequency of type of appointment

df['typeofappointment'].value_counts()
```

```python
# Treatment Prescriptions
# Looking at which medications were prescribed for which diseases

treatment_prescriptions = df.groupby(['diseasediagnosed', 'medicationprescribed']).size().reset_index(name='count')
treatment_prescriptions
```

Use pivot tables or cross-tabulations for multi-dimensional analysis

```python
# Pivot table based on gender, averagebp , and weight (kg) using the mean

df.pivot_table(index='gender', columns='averagebp', values='weightinkg', aggfunc='mean')
```

```python
# Pivot table based on disease diagnosed and medication prescribed

treatment_prescriptions.pivot_table(index='diseasediagnosed', columns='medicationprescribed', values= 'count')
```

Compare Load Times

```python
dataset_path = 'https://raw.githubusercontent.com/zf81/datasci_2_manipulation/main/datasets/healthcare_data_cleaning.csv'
```

```python
# Bringing in dataset with Pandas
start_time = time.time()
pandas_df = pd.read_csv(dataset_path)
pandas_time = time.time() - start_time
```

```python
# Bringing in dataset with Polars
start_time2 = time.time()
polars_df = pl.read_csv(dataset_path)
polars_time = time.time() - start_time2
```

```python
# Display load times
print(f'Pandas load time: {pandas_time:.4f} seconds', pandas_time)
print(f'Polars load time: {polars_time:.4f} seconds', polars_time)
```

Observation: Polars load time for the dataset was faster than Pandas
